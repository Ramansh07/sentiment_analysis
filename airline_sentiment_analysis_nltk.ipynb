```python
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
from nltk import pos_tag
```

```python
from nltk.corpus import stopwords
import string 
stops = set(stopwords.words('english'))
punctuations = list(string.punctuation)
stops.update(punctuations)
```

```python
def get_simple_pos(pos):
    if pos.startswith('J'):
        return wordnet.ADJ
    elif pos.startswith('V'):
        return wordnet.VERB
    elif pos.startswith('R'):
        return wordnet.ADV
    elif pos.startswith('N'):
        return wordnet.NOUN
    else:
        return wordnet.NOUN
```

```python
def clean(words):
    output = []
    for word in words:
        if word.lower() not in stops:
            pos = pos_tag([word])
            w = lemmatizer.lemmatize(word,get_simple_pos(pos[0][1]))
            output.append(w)
    return output
```

```python
clean_df = [(clean(text),category) for text,category in df]
clean_df = np.array(clean_df)
```

```python
len(clean_df)
```

```python
x_train=clean_df[:8000,:]
x_test =clean_df[8000:,0]
y_test = clean_df[8000:,1]
```

```python
all_words = []
for doc in x_train:
    all_words = all_words+doc[0]
print(len(all_words))
```

```python
import nltk
freq = nltk.FreqDist(all_words)
a = freq.most_common(3000)
a
```

```python
features = []
for i in range(len(a)):
    features.append(a[i][0])
```

```python
def get_features(words):
    word_set = set(words)
    current_features = {}
    for feature in features:
        current_features[feature]=feature in word_set
    return current_features
```

```python
training_data = [(get_features(words),category) for words,category in x_train]
testing_data = [(get_features(words)) for words in x_test]
```

```python
from nltk import NaiveBayesClassifier
```

```python
classifier = NaiveBayesClassifier.train(training_data)
```

```python
classifier.classify(testing_data[0])
```

```python
y_pred = []
for i in range(len(testing_data)):
    y_pred.append(classifier.classify(testing_data[i]))
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_pred,y_test)
```
